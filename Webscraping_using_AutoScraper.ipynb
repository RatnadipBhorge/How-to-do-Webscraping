```python
## Lest import AutoScraper
from autoscraper import AutoScraper
```

```python
## search the website url you want to do web scrapping and copy the url
## here I used amazon website for smartphones
url='https://www.amazon.in/s?k=smartphones&ref=nb_sb_noss_2'
```

```python
## Lets prepare the content you want 
wanted_list=['₹6,999','Redmi 9A (Nature Green, 2GB RAM, 32GB Storage) | 2GHz Octa-core Helio G25 Processor | 5000 mAh Battery']
```

```python
## use our function and try to build using .build and print the result 
scraper=AutoScraper()
result=scraper.build(url,wanted_list)
print(result)
```

```python
## we have to use grouped each entity using grouped=True
scraper.get_result_similar(url,grouped=True)
```

```python
## apply aliase for the entity. There is rule  for the every entity eg for price rule_ewr8 and save the scraping
scraper.set_rule_aliases({'rule_ewr8':'price','rule_4pqo':'title'})
scraper.keep_rules(['rule_ewr8','rule_4pqo'])
scraper.save('amazon.search')
```

```python
d=scraper.get_result_similar('https://www.amazon.in/s?k=smartphones&ref=nb_sb_noss_2',group_by_alias=True)
```
